```python
def mod_binary(g,k,n):
    # out: g^k mod n
    bk = bin(k)[2:]
    y = 1
    for i in bk:
        y = y * y % n
        if i == '1':
            y = y * g % n
    return y
```

```python
def rsaEnc(m, n, e):
    return mod_binary(m, e, n)
```

```python
def rsaDec(c, n, d):
    return mod_binary(c, d, n)
```
