```python
def mod_binary(g,k,n):
    # out: g^k mod n
    bk = bin(k)[2:]
    y = 1
    for i in bk:
        y = y * y % n
        if i == '1':
            y = y * g % n
    return y
```

```python
def euclid(a, b):
    # use extended euclid
    # out: gcd(a,b)
    x0, x1 = 1, 0
    y0, y1 = 0, 1
    while b!=0:
        q, r = divmod(a, b)
        x0, x1 = x1, x0 - q * x1
        y0, y1 = y1, y0 - q * y1
        a, b = b, r
    return a 
```

```python
def fermat_test(n, a):
    if mod_binary(a, n-1, n) != 1:
        return False
    return True
```

```python
def fermat_primegen(iv, k):
    cur = 0
    flag = True
    for a in range(2, iv//2 +1):
        if euclid(iv, a):
            cur +=1
            if fermat_test(iv, a) == False:
                flag = False
                break
        if cur == k:
                break
    if flag == True:
        return iv
    else:
        return fermat_primegen(iv+1, k)
```

```python
def prime_arr_gen(bit_size, ret_size, f):
    #     bit_size : 生成素数のビット数
    #     ret_size : 生成素数の個数
    #     f : 判定回数
    cur = 2**(bit_size-1)
    ret = []
    for i in range(ret_size):
        p = fermat_primegen(cur, f)
        cur = p+1
        ret.append(p)
        if (len(str(bin(p)[2:])) > bit_size):
            return ret
    return ret
```

```python
import sympy
def count_prime(arr):
    cnt = 0 
    for x in arr:       
        if sympy.isprime(x) == True:
            cnt +=1 
    return cnt
```

```python
from tqdm import tqdm_notebook as tqdm

bit_size = [128, 256, 512, 756]
counts = {}
for x in bit_size:
    counts[x] = []
    for i in tqdm(range(1, 6)):
        primes = prime_arr_gen(x, 100, i)
        counts[x].append(count_prime(primes))
print(counts)
```
